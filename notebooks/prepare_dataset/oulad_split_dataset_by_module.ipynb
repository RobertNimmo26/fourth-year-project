### Load dataset and split by module

```python
if 'google.colab' in sys.modules:
  path = '../OULAD dataset'
  write_path = 'Deadline'
else:
  path = '../../dataset/Deadline'
  write_path = path
```

```python
deadline_list = [i for i in range(0,271,10)]
```

```python
for deadline in deadline_list:
  oulad_dataset_combined_df = pd.read_csv(f'{write_path}/{deadline}/oulad_dataset_combined_{deadline}.csv')
  modules = oulad_dataset_combined_df.code_module.unique()
  for module in modules:
    module_df = oulad_dataset_combined_df.loc[oulad_dataset_combined_df['code_module'] == module]
    print(f"Module {module}")
    print(module_df['code_presentation'].value_counts())
    print("")
    module_df.drop('code_module', axis=1)
    module_df.to_csv(f'{write_path}/{deadline}/oulad_{module}_{deadline}.csv', index=False)
```
